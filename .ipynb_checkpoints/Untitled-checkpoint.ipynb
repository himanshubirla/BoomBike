<h2> Problem Statement</h2>

A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. This bike can then be returned to another dock belonging to the same system.

BoomBikes has recently suffered considerable dips in their revenues due to the ongoing Corona pandemic. The company is finding it very difficult to sustain in the current market scenario. So, it has decided to come up with a mindful business plan to be able to accelerate its revenue as soon as the ongoing lockdown comes to an end, and the economy restores to a healthy state

Understand the demand for shared bikes among the people after this ongoing quarantine situation ends across the nation due to Covid-19.

Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. The company wants to know:

    -Which variables are significant in predicting the demand for shared bikes.
    -How well those variables describe the bike demands
    
<h3> Things to note </h3>

    - You can observe in the dataset that some of the variables like 'weathersit' and 'season' have values as 1, 2, 3, 4 which have specific labels associated with them. - There is no order associated.
    
    - You might notice the column 'yr' with two values 0 and 1 indicating the years 2018 and 2019 respectively. since these bike-sharing systems are slowly gaining popularity, the demand for these bikes is increasing every year proving that the column 'yr' might be a good variable for prediction.
    
    - "cnt" is target variable. cnt=casual+registered.
    
 

<h2>Initial Hypothesis</h2>

    -During rainy and winter season demand for bikes are less
    -On Holidays demand for casual users demands will be more 
    -On weekdays registered users demands will be more
    -On working days demand for bike will be more 
    -As the weather situation increases from 1 to 4 demand of bike decreases
    - In extreme hot/extreme cold/Heavy rain demand of bike will be less
    -As humidity lowers you shall see increase in demand of bike

```python
def IQR(x):
    return x.quantile(.75)-x.quantile(.25)
def ninetynine_Perc(x):
    return x.quantile(.99)
def Range(x):
    return x.max()-x.min()
```

```python
def analysis(df,col,grby_col):
    print(df[col].describe())
    print('\n')
    print("IQR of {0} {1}".format(col,IQR(df[col])))
    print('\n')
    print(df.groupby(grby_col).agg({col:['describe',IQR,ninetynine_Perc,Range,'skew']}))

    plt.figure(figsize=(20,15))
    plt.subplot(2,2,1)
    plt.title("Distribution of {0}".format(col))
    sns.distplot(df[col],hist=True,kde=True,color='r')
    plt.xlabel("{0}".format(col))
    plt.ylabel("Distribution")
    plt.subplot(2,2,2)
    plt.title("Boxplot of {0}".format(col))
    sns.boxplot(x=col,data=df,color='m')
    plt.xlabel("{0}".format(col))
    plt.ylabel("Distribution")
    plt.subplot(2,2,3)
    plt.title("Boxplot of {0} by {1}".format(col,grby_col))
    sns.boxplot(y=col,x=grby_col,data=df,color='m')
    plt.xlabel("{0}".format(col))
    plt.ylabel("Distribution")
    plt.subplot(2,2,4)
    plt.title("Distribution of {0} by {1}".format(col,grby_col))
    sns.kdeplot(df[col],hue=df[grby_col])
    plt.xlabel("{0}".format(col))
    plt.ylabel("Distribution")
    plt.show()
```

```python
#Analysis of Season
analysis(df,'cnt','season')
```

- As the season approaches summer and fall the demand for bike increases and reaches maximum. There is not much difference in demnad in the two season.

- Demand of bike in rest of seasons appart from spring is way higher. Demand in spring being the lowest. Which means bike company does well significantly in summer and fall.

-  We see as the season changes from spring to summer to fall average bike rental also increases. We also see trend of third quantile increasing , there is sharp increase in third quantile from spring to summer and then there is steady increae in fall , and same is seen for first quantile which means that, in general there are higher rentals during summer and fall, having said that there is opportunity for higher rentals during these season.

- Season does explain the trend in the demand of bike , it does make an inverted U shape.
